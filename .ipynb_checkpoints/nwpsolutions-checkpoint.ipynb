# DMD

Compute DMD modes and predict.

```python
def svd(x):
    U, S, Vh = np.linalg.svd(x, full_matrices=False)
    V = Vh.conj().T
    S = np.diag(S)
        
    return U, S, V
```

```python
def dmd(x, U, S, V, x_prime, r, sort=True):
    U_r = U[:, :r]
    S_r = S[:r, :r]
    V_r = V[:, :r]
    
    A_tilde = U_r.T @ x_prime @ V_r @ np.linalg.inv(S_r)
    
    Lambda, W = np.linalg.eig(A_tilde)
    
    Phi = x_prime @ V_r @ np.linalg.inv(S_r) @ W
    
    if sort:
        sort_indices = np.argsort(np.abs(Lambda))[::-1]
        Lambda = Lambda[sort_indices]
        Phi = Phi[:, sort_indices]
    
    return Phi, Lambda
```

```python
def dmd_pred(X_t, num_modes, num_prev):
    X = np.array(X_t).T
    num_prev = 0 if X.shape[1] < num_prev else num_prev
    x = X[:, -num_prev:-1]
    x_prime = X[:,  -num_prev+1:]
    
    r = min(num_modes, x.shape[1])
    print(x.shape, x_prime.shape, r)
    
    U, S, V = svd(x)
    Phi, Lambda = dmd(x, U, S, V, x_prime, r)
    x_new = X[:, -1]

    # Discrete-time amplitude, if continuous-time is needed (refer to book).
    x_pred = Phi @ np.diag(Lambda) @ (np.linalg.pinv(Phi) @ x_new)
    
    return x_pred
```

# Modified GMRES

This GMRES function is a modified implementation of the one found in SciPySparseLinalg. Curretly its modification exposes Q/H. Can be expanded upon.

*Refer to the standard [scipy.sparse.linalg.gmres](https://https://docs.scipy.org/doc/scipy-1.14.0/reference/generated/scipy.sparse.linalg.gmres.html) for usage/documentation.*

*Helper functions.*

```python
def compute_ML(X, B, B_i, i, n):
    y_i = np.linalg.pinv(np.array(B[i-n:]).T) @ B_i
    return np.array(X[i-n:]).T @ y_i

def computer_rand_ML(X, B, B_i, n, prev):
    rand_chosen = np.random.choice(prev, n)
    y_i = np.linalg.pinv(np.array(B)[rand_chosen].T) @ B_i
    return np.array(X)[rand_chosen].T @ y_i

def to_prob_dist(sim, i, n):
    arr = np.array(list(sim[f'{i}'].items()), dtype=np.float32)
    sorted_indices = np.argsort(arr[:, 0])
    sorted_arr = arr[sorted_indices]
    inv_arr = 1 / sorted_arr[i-n:, 1] # Invert the metrics to make lower scores better.
    return inv_arr / np.sum(inv_arr)

def apply_relative_sim_ML(X, B, B_i, X_prob_dist, B_prob_dist, n):
    X_prob, B_prob = np.array(X[i-n:]).T * X_prob_dist, np.array(B[i-n:]).T * B_prob_dist
    return compute_ML(X_prob.T, B_prob.T, B_i, i, n)

def apply_relative_additive(X, X_prob_dist, i, n):
    return np.sum(np.array(X[i-n:]).T * X_prob_dist, axis=1)
```

```python
def exec_single(a_lookup, b_lookup, ite=None, rtol=1e-5):
    convergence_time = []
        
    X = []
    B = []
    
    for i in range(config.num):
        if config.debug: start = time.time()
        current_history = []
        
        def callback(res_norm):
            current_history.append(res_norm)
        
        try:
            A_i, B_i = sp.load_npz(a_lookup[f'{config.a_head}{i}']), np.load(b_lookup[f'{config.b_head}{i}'])
            x_init_i = np.zeros_like(B_i) if i == 0 or not config.use_prev else X[i-1]

            if config.use_ML:
                if config.static_ML and i > 0:
                    print(f'Computing fixed selection of {i}: {ite} previous.')
                    if i > ite-1 and ite != 0:
                        x_init_i = compute_ML(X, B, B_i, i, ite)
                    elif ite == 0:
                        x_init_i = compute_ML(X, B, B_i, i, i)
                            
                elif config.dynamic_ML:
                    if config.converge_criteria and i > 0:
                        print(f'Computing convergence criteria of {i}: {config.conv_n}, use previous ({config.fixed_last}), use 4 previous ({config.fixed_last_4}).')
                        good_sols = np.array(convergence_time) < config.conv_n
                        print(np.sum(good_sols))
                        if np.sum(good_sols) == 0 and i < 4:
                            x_init_i = X[i-1]
                        else:
                            if config.fixed_last:
                                good_sols[-1] = True  
                            if config.fixed_last_4 and i > 3:
                                good_sols[-4], good_sols[-3], good_sols[-2], good_sols[-1] = True, True, True, True

                            X_sel = np.array(X)[good_sols, :]
                            B_sel = np.array(B)[good_sols, :]
                                                        
                            x_init_i = compute_ML(X_sel, B_sel, B_i, i, i)
                            
                    elif config.random_selection:
                        print(f'Computing random selection of {i}: {ite} previous.')
                        if i > ite-1 and ite != 0:
                            computer_rand_ML(X, B, B_i, i, ite)
                        elif ite == 0:
                            computer_rand_ML(X, B, B_i, i, i)
                                
                    elif config.relative_sim and i > 0:
                        print(f'Computing relative sim of {i}: {ite} previous.')
                        if i > ite-1 and ite != 0:
                            B_prob_dist, X_prob_dist = to_prob_dist(config.b_sim, i, ite), to_prob_dist(config.x_sim, i, ite)
                            x_init_i = apply_relative_additive(X, X_prob_dist, i, ite)
                        elif ite == 0:
                            B_prob_dist, X_prob_dist = to_prob_dist(config.b_sim, i, i), to_prob_dist(config.x_sim, i, i)
                            x_init_i = apply_relative_additive(X, X_prob_dist, i, i)
                            
                    elif config.dmd and i > 1:
                        if config.ite != 0 and config.r_modes > config.ite:
                            print('Cannot use more modes than previous history.')
                            break
                        print(f'Computing DMD of {i}: {config.r_modes} modes, {ite} previous.')
                        x_init_i = dmd_pred(X, config.r_modes, config.ite)
                    
                    
            x_sol, exit_code = solvers.modified_gmres(A_i, B_i, x_init_i, rtol=rtol, callback=callback, callback_type='pr_norm')
            print(len(current_history))
            convergence_time.append(len(current_history))
            
            X.append(x_sol)
            B.append(B_i)

            if exit_code != 0:
                print(f'Error on timestep {i}.')

            if config.debug: 
                end = time.time()
                print(end-start)
                
            print(f'Timestep {i} done.')
            
        except Exception as e:
            print(f"Exception occurred at timestep {i}: {e}")
            
    return convergence_time
```

```python
def exec_expirement01(a_lookup, b_lookup, n):
    convergence_time = []
    residuals = []
    n = n
    
    def callback(res_norm):
        current_history.append(res_norm)
    
    def to_prob_dist(sim, i, n):
        arr = np.array(list(sim[f'{i}'].items()), dtype=np.float32)
        sorted_indices = np.argsort(arr[:, 0])
        sorted_arr = arr[sorted_indices]
        inv_arr = 1 / sorted_arr[i-n:, 1] # Invert the metrics to make lower scores better.
        return inv_arr / np.sum(inv_arr)
    
    def apply_relative_additive(X, X_prob_dist, i, n):
        return np.sum(np.array(X[i-n:]).T * X_prob_dist, axis=1)
        
    X = []
    B = []
        
    for i in range(config.num):
        if config.debug: start = time.time()
        current_history = []
        
        try:
            A_i, B_i = sp.load_npz(a_lookup[f'{config.a_head}{i}']), np.load(b_lookup[f'{config.b_head}{i}'])
            x_init_i = np.zeros_like(B_i) 
                        
            if i > n-1:
                X_prob_dist =  to_prob_dist(config.b_sim, i, n)        
                x_init_i = apply_relative_additive(X, X_prob_dist, i, n)           
            
            x_sol, exit_code = solvers.modified_gmres(A_i, B_i, x_init_i, callback=callback, callback_type='pr_norm', rtol=1e-5)
                
            X.append(x_sol)
            B.append(B_i)
            
            print(len(current_history))
            convergence_time.append(len(current_history))

            if exit_code != 0:
                print(f'Error on timestep {i}.')

            if config.debug: 
                end = time.time()
                print(end-start)
                
            print(f'Timestep {i} done.')
            
        except Exception as e:
            print(f"Exception occurred at timestep {i}: {e}")
            
    return convergence_time
```

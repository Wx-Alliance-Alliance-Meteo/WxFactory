```python
def get_num_files(directory):
    count = 0

    for filename in os.listdir(directory):
        f = os.path.join(directory, filename)

        # checking if it is a file
        if os.path.isfile(f) and config.a_head == filename[:len(config.a_head)]:
            count += 1

    return count
```

## Asynchronous Cache

Parallel/asynchronous reading in of datasets.

```python
def load_data(key, lookup, header, cache, tail=''):
    if header == config.a_head:
        cache[key] = sp.load_npz(lookup[f'{header}{key}{tail}'])
    else:
        cache[key] = np.load(lookup[f'{header}{key}{tail}'])
        
    if config.debug:
        print(f'\"{header}-{key}-{tail}\" loaded.')

def async_cache_dataset(a_lookup, b_lookup, x_lookup, q_lookup, h_lookup, r_lookup):
    a_cache = {}
    b_cache = {}
    x_cache = {}
    q_cache = {}
    h_cache = {}
    r_cache = {}

    with concurrent.futures.ThreadPoolExecutor() as executor:
        futures = []
        for i in range(config.num):
            if config.a_bool:
                futures.append(executor.submit(load_data, i, a_lookup, config.a_head, a_cache))
            if config.q_bool:
                futures.append(executor.submit(load_data, i, q_lookup, config.q_head, q_cache, config.q_tail))
            if config.h_bool:
                futures.append(executor.submit(load_data, i, h_lookup, config.h_head, h_cache, config.h_tail))
            if config.b_bool:
                futures.append(executor.submit(load_data, i, b_lookup, config.b_head, b_cache))
            if config.x_bool:
                futures.append(executor.submit(load_data, i, x_lookup, config.x_head, x_cache))
            if config.r_bool:
                futures.append(executor.submit(load_data, i, r_lookup, config.r_head, r_cache))
                
        concurrent.futures.wait(futures)
    
    return (a_cache, b_cache, x_cache, q_cache, h_cache, r_cache)
```

```python
config.full_flag = False
config.save_similarities = True
config.debug = True
```

```python
def compute_similarity():
    a_sim = {}
    b_sim = {}
    x_sim = {}
    q_sim = {}
    h_sim = {}
        
    for i in range(config.num):
        ia_output = {}
        ib_output = {}
        ix_output = {}
        iq_output = {}
        ih_output = {}
        
        for k in range(config.num if config.full_flag else i):
            if i == k:
                continue
            
            if config.a_bool:
                A_i, A_k = a_cache[i], a_cache[k]
                ia_output[k] = sp.linalg.norm(A_i - A_k) / sp.linalg.norm(A_k) # Calculate A similarity.
            if config.b_bool:
                b_i, b_k = b_cache[i], b_cache[k]
                ib_output[k] = la.norm(b_i - b_k) / la.norm(b_k) # Calculate b similarity.
            if config.x_bool: 
                x_i, x_k = x_cache[i], x_cache[k]
                ix_output[k] = la.norm(x_i - x_k) / la.norm(x_k) # Calculate x similarity.
            if config.q_bool:
                q_i, q_k = q_cache[i], q_cache[k]
                iq_output[k] = la.norm(q_i - q_k) / la.norm(q_k) # Calculate Q similarity.
            if config.h_bool:
                h_i, h_k = h_cache[i], h_cache[k]
                ih_output[k] = sp.linalg.norm(h_i - h_k) / sp.linalg.norm(h_k) # Calculate H similarity.
                
        a_sim[i] = ia_output
        b_sim[i] = ib_output
        x_sim[i] = ix_output
        q_sim[i] = iq_output
        h_sim[i] = ih_output
        
        if config.debug:
            print(f'Similarities calculated for {i}.')
            
    return a_sim, b_sim, x_sim, q_sim, h_sim
```

```python
def display_plots_granularity(n, title, sim):

    fig, ax = plt.subplots(nrows=n//2, ncols=2, figsize=(20, 8*(n//2)))
    if not config.save_plots:
        fig.suptitle(title)

    def plot_range(start, end, ax, title):
        for i in range(start, end):
            ax.plot(sim[i].keys(), sim[i].values())
        ax.set_title(f'{title}: {start}-{end}')
        
    range_size = config.num // n
    
    for i in range(n):
        plot_range(i*range_size, (i+1)*range_size, ax[i//2, i%2], title)
        
    if config.save_plots:
        plt.savefig(f"{''.join(title.split(' '))}{n}.png")
```

# Experiment 2

Look at Krylov subspaces.

```python
def do_they_span(q1, q2, ones):
    # Orthogonalize Q1 and Q2.
    q1_orth = la.orth(q1)
    q2_orth = la.orth(q2)

    # Transpose of Q1.
    qt = q1_orth.T

    # Matrix multiplication.
    qc = np.dot(qt, q2_orth)

    # Singular value decomposition.
    _, svd, _ = np.linalg.svd(qc)

    return np.allclose(svd, ones, rtol=10e-7)
```

```python
def test_span():
    ones = ones = np.ones(q_cache[0].shape[0])
    confusion = np.zeros((len(q_cache), len(q_cache)))
        
    for i in range(config.num):
        for k in range(i+1, config.num):
            q1, q2 = q_cache[i], q_cache[k]
            confusion[i, k] = 1 if do_they_span(q1, q2, ones) else 0
            
        if config.debug:
            print(f'Spans computed for {i}.')
            
    return confusion
```
